```python
df1_2 = pd.read_csv('df1_2.csv', header=0)
df2_2 = pd.read_csv('df2_2.csv', header=0)
```

Para la normalización del presente conjunto de datos, se utilizará el algoritmo Min-Max Scaling.

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# Suponiendo que tienes df1_2 y df2_2 ya cargados

# Combina los conjuntos de datos df1_2 y df2_2
combined_data = np.vstack((df1_2, df2_2))

# Crea una instancia del escalador MinMaxScaler
scaler = MinMaxScaler()

# Ajusta y transforma el conjunto de datos combinado
combined_data_normalized = scaler.fit_transform(combined_data)

# Ahora, separa los datos normalizados nuevamente en df1_3 y df2_3 y crea DataFrames
df1_3 = pd.DataFrame(combined_data_normalized[:len(df1_2)], columns=df1_2.columns)
df2_3 = pd.DataFrame(combined_data_normalized[len(df1_2):], columns=df2_2.columns)

# Verifica las dimensiones de los DataFrames normalizados si es necesario
print("df1_3 shape:", df1_3.shape)
print("df2_3 shape:", df2_3.shape)
```
