## Imputación de data faltante con algoritmo KNN

Al tener variables anonimizadas, no se puede discriminar si los outliers presentan valores lógicos, por lo que para resguardar la integridad de la data, se considera mantenerlos.

Imputación de valores vacíos con kNN.

Para el correcto funcionamiento del algoritmo, se cambian los valores de la columna class por 0 y 1 para neg y pos, respectivamente.

```python
# Reemplazar "neg" por 0 y "pos" por 1 en la columna "class" de df1_1
df1_1['class'] = df1_1['class'].replace({'neg': 0, 'pos': 1})

# Reemplazar "neg" por 0 y "pos" por 1 en la columna "class" de df2_1
df2_1['class'] = df2_1['class'].replace({'neg': 0, 'pos': 1})
```

En el seteo del algoritmo, se utilizan 5 vecinos, y los pesos se distribuyen de manera uniforme.

```python
imputer = KNNImputer(n_neighbors=5, weights="uniform")
df1_2 = imputer.fit_transform(df1_1)
df2_2 = imputer.fit_transform(df2_1)

```

la salida del imputador entrega los conjuntos de datos como un arreglo, y no como un data frame.
Para corregirlo, se utiliza una función que transforma los arreglos a data frames.

```python
import pandas as pd

def array_to_dataframe(data_array, column_names):
    # Crear un DataFrame a partir del array y los nombres de columna proporcionados
    df = pd.DataFrame(data_array, columns=column_names)
    return df

# Ejemplo de uso
data_array = df1_2  # Reemplaza df1_2 con tu array real
column_names = [df1_1.columns]  # Reemplaza con los nombres de columna correspondientes

# Convertir el array en DataFrame
df1_2 = array_to_dataframe(data_array, column_names)

# Ejemplo de uso
data_array = df2_2  # Reemplaza df1_2 con tu array real
column_names = [df2_1.columns]  # Reemplaza con los nombres de columna correspondientes

# Convertir el array en DataFrame
df2_2 = array_to_dataframe(data_array, column_names)

```

## Normalización data

Al no eliminar los outliers, una manera de reducir su influencia sobre las predicciones de los algoritmos de clasificación es la normalización de las catacterísticas. Mejorando la convergencia y rendimiento general de los modelos. También ayuda en la eliminación de los sesgos, mientras mantiene la información relativa entre características.

Para la normalización del presente conjunto de datos, se utilizará el algoritmo Min-Max Scaling.

```python
# Obtener las columnas a escalar sin incluir la variable "class"
#columns_to_scale = df1_1.drop('class', axis=1).columns

# Crear el objeto MinMaxScaler
scaler = MinMaxScaler()

# Ajustar el scaler con los datos de entrenamiento y transformar los datos de entrenamiento
df1_3 = pd.DataFrame(scaler.fit_transform(df1_2.values), columns=df1_2.columns)

# Transformar los datos de prueba utilizando el scaler ajustado con los datos de entrenamiento
df2_3 = pd.DataFrame(scaler.transform(df2_2.values), columns=df2_2.columns)

```
