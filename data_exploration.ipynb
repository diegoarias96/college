**En ambos gráficos se puede ver la presencia de outliers muy por encima del cuartil 99. La técnica utilizada para tratar estos fenómenos será el establecer un límite superior e inferior para truncar los valores de cada característica. El límite superior será el percentil 99, mientras que el límite inferior será el percentil 1.**

```python
# Obtener las columnas desde la columna 1 en adelante
columns_to_process = df1.columns[1:]

# Aplicar la truncación de outliers en df1 para todas las columnas desde la columna 1 en adelante
for column in columns_to_process:
    lower_limit = df1[column].quantile(0.01)  # Límite inferior (percentil 1)
    upper_limit = df1[column].quantile(0.99)  # Límite superior (percentil 99)
    df1[column] = df1[column].apply(lambda x: lower_limit if x < lower_limit else x)
    df1[column] = df1[column].apply(lambda x: upper_limit if x > upper_limit else x)

# Aplicar la truncación de outliers en df2 para todas las columnas desde la columna 1 en adelante
for column in columns_to_process:
    lower_limit = df2[column].quantile(0.01)  # Límite inferior (percentil 1)
    upper_limit = df2[column].quantile(0.99)  # Límite superior (percentil 99)
    df2[column] = df2[column].apply(lambda x: lower_limit if x < lower_limit else x)
    df2[column] = df2[column].apply(lambda x: upper_limit if x > upper_limit else x)
```

Una variable objetivo normalmente distrivuida (o cerca), ayuda en el mejor modelaje de relaciones entre el objetivo y las variables independientes.

```python
correlation_train = df1.corr()
corr_dict = correlation_train['class'].sort_values(ascending=False).to_dict()
important_columns1=[]
for key,value in corr_dict.items():
    if ((value>0.2) & (value<0.9)) | (value<=-0.2):
        important_columns1.append(key)
        
cant = len(important_columns1)

print("Cantidad de valores:", cant)        
print("correlated columns ",important_columns1)

```

```python
correlation_train = df2.corr()
corr_dict = correlation_train['class'].sort_values(ascending=False).to_dict()
important_columns2=[]
for key,value in corr_dict.items():
    if ((value>0.2) & (value<0.9)) | (value<=-0.2):
        important_columns2.append(key)
        
cant = len(important_columns2)

print("Cantidad de valores:", cant)        
print("correlated columns ",important_columns2)
```
